```python
import pandas as pd
import os
```

#### Merge 12 months of sales data into a single CSV file

```python
#Show all the csv files have in folder Sales_Data for analysis

files = [file for file in os.listdir('./Sales_Data')]

for file in files:
    print(file)
    
```

```python
#Create empty DataFrame to read the csv files data

all_months_data = pd.DataFrame()
```

```python
files = [file for file in os.listdir('./Sales_Data')]

for file in files:
    df=pd.read_csv('./Sales_Data/'+file)
    all_months_data =pd.concat([all_months_data,df])

#save merged data as all_data to single csv file,create csv file in Sales_Data folder .    
all_months_data.to_csv("all_data.csv" ,index=False)  
```

#### read updated Dataframe

```python
all_data=pd.read_csv('all_data.csv')

all_data.head()
```

```python
all_data.info()
```

#### Clean Up The Data from NAN Values.

```python
#Find Columns with NAN 

nan_df=all_data[all_data.isna().any(axis=1)]

nan_df.head()
```

```python
#Drop NAN rows
all_data=all_data.dropna(how='all')
all_data.head()
```

```python
# got error when try to convert month 04 string to int numbers ,*ValueError: invalid literal for int() with base 10: 'Or'
```

#### Find 'Or' and delete it

```python
temp_df=all_data[all_data['Order Date'].str[0:2]=='Or']

temp_df.head()
```

```python
#get new all_data without 'Or' Values

all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
```

#### Convere columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])# Make float

all_data.info()
```

###now we can add month column

#### Add specific column for the month


```python
#Add month column
all_data['Month']= all_data['Order Date'].str[0:2]
all_data['Month']= all_data['Month'].astype('int32')

all_data.head()
```

#### Add Sales column

```python
#Multiply 'Quantity Ordered' column With 'Price Each' colum

all_data['Sales']=all_data['Quantity Ordered'].astype('int') * all_data['Price Each'].astype('float')

all_data.head()
```

### * What city had the highest number of sales ?

#### Add 'City' column

```python
all_data.head()
```

```python
# Use .apply() method and split the city in 'Purchase Address' column,[1]~index1 in 'Purchase Address',[2]~index2

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1] #Add (State code)

all_data['City']=all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}({get_state(x)})")

all_data.head()
```

```python
results=all_data.groupby('City').sum()
results
```

```python
cities=[city for city, df in all_data.groupby('City')]

plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation='vertical',size=8) #rotate x~names to vertical
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')
plt.show()
```

* San Francisco had the highest number of sales 8.26M

### *What time should we display advertisements to maximize likelihood of customers buying product ?

```python
all_data.head()
```

#### Convert 'Order Date' column to DateTime format

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

#### Add 'Hour' and 'Minute' columns

```python
all_data['Hour']=all_data['Order Date'].dt.hour

all_data['Minute']=all_data['Order Date'].dt.minute

all_data.head()
```

```python
hours=[hours for hours, df in all_data.groupby('Hour')]

plt.plot(hours,all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

* around 12:00 am and 19:00 pm we have a pick when people made orders, between 11-12 or 18-19 best time to post ad

### What products are most often sold together ?

```python
all_data.head()
```

* If the product have the same 'Order ID' so they Sold together

```python
df= all_data[all_data['Order ID'].duplicated(keep=False)]

df.head(5)
```

#### Add new column 'Grouped'

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df=df[['Order ID','Grouped']].drop_duplicates() #drop the same pairs 

df.head()
```

#### Count the pairs 

```python
from itertools import combinations
from collections import Counter
```

```python
count=Counter()

for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2))) #rows of~2 products 
    
for key, value in count.most_common(10):
    print(key,value)
```

```python
prices=all_data.groupby('Product').mean()['Price Each']

print(prices)
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```

* The quantity order high when the price is low , this is the case for the best selling products !
